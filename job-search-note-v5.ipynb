```typescript
import { ChatPromptTemplate } from "npm:@langchain/core/prompts";
import { StringOutputParser } from "npm:@langchain/core/output_parsers";
import { ChatOpenAI } from "npm:@langchain/openai";
import { config } from "https://deno.land/x/dotenv/mod.ts";

const env = config({ path: "./.env", export: true });

const model = new ChatOpenAI({
  apiKey: env.OPENAI_API_KEY,
  modelName: "gpt-4-1106-preview",
});

const parser = new StringOutputParser();

const systemTemplate = `
"""You are an expert and especially trained machine model to recognize the given job query and constructing a query string for an API

------ Instructions ------
1. Read the given job query string and identify the keywords or their meaning to be used as query parameter values.
2. Find the location, date posted, employment type, remote job status, and salary from the query.
3. Construct a query string for an API with the recognized values, remember the job_title should be in lowercase.
4. If the job query includes salary information, exclude the salary from the query string.
5. The query string should be constructed in the following format:
    - q=job_title&location=location&chips=date_posted:option,employment_type:option&hl=en&Itype=option
6. Return only the constructed query string, no other information or text should be included.
    
------ Accepted Query Parameter Values ------
1. location: This can be any location such as a country, city, or street, when given is a country always expand the country name (UK -> united kingdom). If not mentioned, exclude the parameter.
2. date_posted: This can be one of the following options:
    - today
    - ndays (where n is a number from 0 to 9)
    - week
    - month
    - years
    If not provided, keep "month" as the default value.
3. employment_type: This can be one of the following options:
    - CONTRACTOR
    - FULLTIME
    - INTERN
    - PARTTIME
    If not provided, exclude the field.
4. Itype: This can be either 1 (for remote job) or 0 (for non-remote job). Exclude if not provided.

#### START EXAMPLES

------ Example Inputs ------
1. Seeking an entry-level digital marketing specialist in New York (remote) offering $2500 posted within a week.
2. Junior software developer job in Germany as part-time.
3. Looking for an entry-level data analyst in Canada (remote) with CAD 4000 monthly salary and should be contract-based.
4. Intern HR coordinator in the UK salary posted within this month and not remote/on-site.
5. Looking for a sales representative job in Japan with a monthly salary of ¥300,000 salary hybrid posted within 3 days.

------ Example Outputs ------
1.  "q=entry+level+digital+marketing+specialist&location=New+York&chips=date_posted:week,employment_type:FULLTIME&hl=en&salary=2500&Itype=1
2.  "q=junior+software+developer&location=Germany&chips=date_posted:month,employment_type:PARTTIME&hl=en
3.  "q=entry+level+data+analyst&location=Canada&chips=date_posted:month,employment_type:CONTRACTOR&hl=en
4.  "q=HR+coordinator&location=UK&chips=date_posted:month,employment_type:INTERN&hl=en&Itype
5.  "q=entry+level+sales+representative&location=Japan&chips=date_posted:3days&hl=en&Itype
#### END EXAMPLES
"""
`;

const promptTemplate = ChatPromptTemplate.fromMessages([
  ["system", systemTemplate],
  ["user", "{jobQuery}"],
]);

const chain = promptTemplate.pipe(model).pipe(parser);

const result = await chain.invoke({ jobQuery });

console.log("result", result);
```

```typescript
import axiod from "https://deno.land/x/axiod/mod.ts";

async function fetchJobListings(
  queryString: string
): Promise<{ job_title: string; employment_type: string }[]> {
  const queryParams = queryString.replace(/["\\]/g, "");
  const url = `https://serpapi.com/search.json?engine=google_jobs&${queryParams}&api_key=${env.SERP_API_KEY}`;

  try {
    const response = await axiod.get(url);

    // Extract the relevant data from the response and format it for the searchJobs function
    return (
      response.data?.jobs_results?.map((job) => ({
        job_title: job.title,
        company_name: job.company_name,
        location: job.location,
        relatedLink: job.related_links[0]?.link || "No related links",
      })) || []
    );
  } catch (error) {
    console.error("Error fetching jobs from SERP API:", error);
    return error.message;
  }
}

async function fetchData(): Promise<string> {
  try {
    const descriptions: string[] = [];

    await Promise.all(
      courseList.flatMap((term) =>
        term.courses
          .filter((course) => course.status === "Completed")
          .map((course) => fetchCourseDescription(course))
      )
    ).then((results) => {
      descriptions.push(...results);
    });

    return descriptions.join("|");
  } catch (error) {
    console.error("Error fetching data:", error);
    return "";
  }
}

async function fetchCourseDescription(course: Course): Promise<string> {
  const { strm, catalog_number, subject, course_title } = course;
  const url = `https://api.myasuplat-dpl.asu.edu/api/course/${strm}/subject/${subject}/${catalog_number}`;

  try {
    const response = await axiod.get(url);
    const [data] = response.data;

    if (!data || !data.fullDescription) {
      throw new Error("Invalid response structure");
    }

    return `${course_title} (${catalog_number}): ${data.fullDescription}`;
  } catch (error) {
    console.error(
      `Error fetching course description for ${catalog_number}:`,
      error
    );
    return `${course_title} (${catalog_number}): Description not available`;
  }
}

const matchedJobs = await fetchJobListings(result);
console.log("Jobs:", JSON.stringify(matchedJobs, null, 2));

const courseDescriptions: string = await fetchData();
console.log("Course descriptions:", courseDescriptions);
```

```typescript
const jobSuggestionTemplate = `You are an expert in analyzing the course descriptions of the courses the user has opted for, and jobs that are available and returning the top 5 job suggestion mached based on similarity and their match.
>>>>>INPUT<<<<<
###
course descriptions: {courseDescriptions},
Jobs: {matchedJobs}
###

Follow the instructions to achieve the best results:
==== INSTRUCTIONS ====
1. For the given course descriptions which is separated by "|", analyze the jobs the user can apply for.
2. Extract jobs titles from the provided of json (jobs - array of objects).
3. Perform an intelligent matching operation to extract the best suitable jobs for the user based on course descriptions.
4. Return the results in a clear format with job title, company name, location and related_link in JSON . Include a concise 2-3 line explanation for each job, describing why it was selected. This explanation should relate the job suitability to the courses taken by the user.
####
Example:-"This job matches your profile because it requires skills developed in your courses, making it a suitable option for your career progression."
####
==== END ====

==== IMPORTANT ====
1. Don’t return any code
2. Limit the response to 50 words
3. Just provide the list of best suitable job titles and short note on why from the given jobs list based on the course descriptions provided.
4. Do not include any jobs from your end.
5. If no suitable job matches are found based on the courses described, first politely apologize and inform the user that no job matches were found and suggest that the user modifies their job query to align more closely with their course description  and give the formatted response with status and message and alternative query suggestion. 
==== END ====

`;

const promptTemplate = ChatPromptTemplate.fromMessages([
  ["system", jobSuggestionTemplate],
]);

const chain = promptTemplate.pipe(model).pipe(parser);

const stringifiedMatchedJobs = JSON.stringify(matchedJobs);

const result = await chain.invoke({
  courseDescriptions,
  matchedJobs: stringifiedMatchedJobs,
});

console.log("result", result);
```
